```python
# Start off for realz, with out imports
import numpy as np
import matplotlib.pyplot as plt # again.... cuz... im g like that
from matplotlib.ticker import FuncFormatter

# todays darta
data = {'Barton LLC': 109438.50,
        'Frami, Hills and Schmidt': 103569.59,
        'Fritsch, Russel and Anderson': 112214.71,
        'Jerde-Hilpert': 112591.43,
        'Keeling LLC': 100934.30,
        'Koepp Ltd': 103660.54,
        'Kulas Inc': 137351.96,
        'Trantow-Barrows': 123381.38,
        'White-Trantow': 135841.99,
        'Will LLC': 104437.60}

# assigning variable names to keys and vals of dic and calculating the mean
data_values = list(data.values())
data_keys = list(data.keys())
data_mean = np.mean(data_values)
```

```python
# Time to plot our interesting data

fig, ax = plt.subplots(figsize=(8,6))

# It makes sense but it is werid that the x and y vales are flipped
ax.barh(data_keys, data_values)

plt.show()
```

```python
# So to format our x axis, which is meant to be dollars, we are going to
# Make a function that will take in the value and position of the value
# format it and return it.

def currency(x, pos):
    """The two args are the value and tick position"""
    if x >= 1e6:
        s = "${:1.1f}M".format(x*1e-6)
    else:
        s = '${:1.0f}K'.format(x*1e-3)
    return s

# FuncFormatter is from the matplotlib library.
formatter = FuncFormatter(currency)

# Personally I didnt write this code, and I dont like it, feels clunkly and old...
```

```python
# We plot again

# All ive been told is that if you have labels or data cut off for some reason
# This lil one liner should fix you up. tidies everything up, tightens it abit
plt.rcParams.update({"figure.autolayout": True})

fig, ax = plt.subplots(figsize=(8,6))

ax.barh(data_keys, data_values)
ax.set_xlabel("Performance metric")
ax.set_ylabel("Company names", rotation=90)
ax.set_title("Company vs performance")

# We are setting a variable to control the tick values on the axis
labels = ax.get_xticklabels()

# Now we are using the set propeties method to edit the labels
plt.setp(labels, rotation=45, horizontalalignment="right")

# Now we add customization for our x and y labels, again uysing the setp method
# This time its just set and its on the ax object
# xlim sets the start and end point for the x axis in terms of values
ax.set(xlim=[-10000,140000], xlabel="Productitivy", ylabel="Comapny name", title="Company v productivity")

# Adding the function for formatting
ax.xaxis.set_major_formatter(formatter)

plt.show()
```

```python
# THE LAST REPEAT OF THIS PLOT 

# All ive been told is that if you have labels or data cut off for some reason
# This lil one liner should fix you up. tidies everything up, tightens it abit
plt.rcParams.update({"figure.autolayout": True})

fig, ax = plt.subplots(figsize=(8,6))

ax.barh(data_keys, data_values)
ax.set_xlabel("Performance metric")
ax.set_ylabel("Company names", rotation=90)
ax.set_title("Company vs performance")

# We are setting a variable to control the tick values on the axis
labels = ax.get_xticklabels()

# Now we are using the set propeties method to edit the labels
plt.setp(labels, rotation=45, horizontalalignment="right")

# We want to add avertical line to our plot that shows us the average production for a company
# To do this we are going to use the axvline function. The data_mean comes from the start of this
# notebook. ls is line style and color is.... nevermind
ax.axvline(data_mean, ls=":", color="m")

# Next we are going to add an abitrary text item to give more context to our data. In particular
# We want to show which company is a new company, we can add arb text with the text function
# first a for loop to iterate through out new companies
for company in [3,5,8]:
    # Starting with where, thats outside our xlim, then the data, and other obvious stuff
    ax.text(145000, company, "New company", fontsize=10, verticalalignment="center")


# Now we add customization for our x and y labels, again uysing the setp method
# This time its just set and its on the ax object
# xlim sets the start and end point for the x axis in terms of values
ax.set(xlim=[-10000,140000], xlabel="Productitivy", ylabel="Comapny name", title="Company v productivity")

# Adding the function for formatting
ax.xaxis.set_major_formatter(formatter)

# One last thing is we are adding values to our xticks
ax.set_xticks([0, 25e3, 50e3, 75e3, 10e4, 12.5e4])

plt.show()
```
