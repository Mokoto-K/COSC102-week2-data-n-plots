```python
# Starting by import what we need
import matplotlib.pyplot as plt
import numpy as np

# And im just creating the most basic plot to begin with
fig, ax = plt.subplots(figsize=(8,4))

# Create some datasets to plot
x_plot = [1,2,3,4]
y_plot = [5,2,6,4]

# Plot our data
ax.plot(x_plot, y_plot, "m")
plt.show()
```

```python
# This is meant to be an exmplae of oop apprach to making plots
import matplotlib.pyplot as plt
import numpy as np

# Returns a list of evenly spaced numbers
x_param = np.linspace(0, 2,100)

# Initializing the plot
fig, ax = plt.subplots()

# Plot three different sets of lines in the same plot
ax.plot(x_param, x_param, label="linear")
ax.plot(x_param, x_param**2, label="Quadratic")
ax.plot(x_param, x_param**3, label="Cubic")

# Adding titles and labels
ax.set_xlabel("x label")
ax.set_ylabel("y label")
ax.set_title("Basic Plot")

# Adding a legend, the label in the ax.plot, populates the legend
ax.legend()
plt.show()
```

```python
# Final example is showing how you put multiple charts on the same figure

# We create some random...... ddaaytaa. Note this is making a 2 dimensional array
# of data with 100 values in each array
np.random.seed(345345345)
dayta = np.random.randn(2,100)

# We create a 2 by 2 plot()
fig, ax=plt.subplots(2,2, figsize=(8,8))

# We access each plot using the indexes
ax[0,0].hist(dayta[0])
ax[0][1].scatter(dayta[0], dayta[1])
ax[1][0].plot(dayta[0], dayta[1])
ax[1,1].hist2d(dayta[0], dayta[1])

plt.show()
```
